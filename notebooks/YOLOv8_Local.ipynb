```python
import torch
from PIL import Image
from IPython.display import display
from ultralytics import YOLO
```

```python
# Charger le modèle YOLOv8 pré-entraîné
model = YOLO('yolov8n.pt')  # 'n' pour Nano, léger et rapide
```

```python
from google.colab import files
from PIL import Image
from IPython.display import display

# Téléverser une image depuis votre ordinateur
uploaded = files.upload()

# Récupérer le chemin de l'image téléversée
image_path = list(uploaded.keys())[0]
image = Image.open(image_path)
display(image)
```

```python
# Prédire les objets dans l'image
results = model(image_path, show=True)  # show=True affiche directement les résultats
```

```python
# Sauvegarder et afficher l'image annotée
annotated_image = results[0].save(filename='output.jpg')
display(Image.open('output.jpg'))
```
